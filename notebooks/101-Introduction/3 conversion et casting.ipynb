# Conversion et Casting

## Implicite

Implicite
Une conversion implicite est l’affectation d’une variable avec une autre valeur de type différent sans indiquer au compilateur le type converti désiré
Ce type de conversion est plus dangereux dû à la possibilité de perte de données

**Exemple :** passer d’un `int` à un `short` va lancer une erreur de compilation tandis que passer d’un `short` à un `int` est accepté.



```
int a = 5;
short b = 0;

b = a + 5;
a = b + 5;
```

## Explicite

Une conversion explicite est l’affectation d’une variable avec une autre valeur de type différent en indiquant au compilateur le type converti désiré

Ce type de conversion est moins dangereux puisque le programmeur indique qu’il peut s’attendre à la perte de données

**Exemple :** passer d’un `int` à un `short` avec une conversion explicite est accepté.



```
int a = 5;
short b = 0;

b = (short)(a + 5);
a = b + 5;

a
```

# Overflow et Underflow

Une opération arithmétique peut causer un « oveflow » ou « underflow » si les limites du type ne sont pas respectées

**Exemple :**
Additionne deux `byte` (`240` et `105`) et store le résultat dans un autre `byte` (`345`)
L’addition donne `89` : `345 - 256` (la valeur maximale d’un `byte` est `256` puisque ses limites sont de `0`-`255`)

```
byte a = 240;
byte b = 105;
//byte c = a + b;
//var c = a + b;
//Console.WriteLine(c.GetType().Name);
byte c = (byte)(a + b);

Console.WriteLine($"a: {a}");
Console.WriteLine($"b: {b}");
Console.WriteLine($"c: a + b = {c}");
```
