# Conversion et Casting

## Implicite

Implicite
Une conversion implicite est laffectation dune variable avec une autre valeur de type diffrent sans indiquer au compilateur le type converti dsir
Ce type de conversion est plus dangereux d  la possibilit de perte de donnes

**Exemple :** passer dun `int`  un `short` va lancer une erreur de compilation tandis que passer dun `short`  un `int` est accept.



```
int a = 5;
short b = 0;

b = a + 5;
a = b + 5;
```

## Explicite

Une conversion explicite est laffectation dune variable avec une autre valeur de type diffrent en indiquant au compilateur le type converti dsir

Ce type de conversion est moins dangereux puisque le programmeur indique quil peut sattendre  la perte de donnes

**Exemple :** passer dun `int`  un `short` avec une conversion explicite est accept.



```
int a = 5;
short b = 0;

b = (short)(a + 5);
a = b + 5;

a
```
