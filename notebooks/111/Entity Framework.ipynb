## ADO.NET en mode EntityFramework

Le mode EntityFramework (EF) de ADO.NET rassemble les avantages des modes connectés et déconnectés :
- L’aspect fortement typé du mode déconnecté
- Des données toujours à jour du mode connecté
- L’utilisation du LINQ du mode déconnecté (LINQ to Entities)
- L’aspect de création automatique de requête SQL du mode déconnecté


## Les entités

L’aspect fortement typé du EF est grâce aux classes fortement typées appelées des entités.

Les entités forment le modèle conceptuel d’une base de données physique à l’intérieur de votre application.

Ces entités peuvent être automatiquement générées par EF, créées par le programmeur ou même modifiées après avoir été créées par EF.

Le EF va automatiquement faire le lien entre les entités et la base de données physique (cette couche est transparente au programmeur).


## Le Entity Data Model (EDM)

Le lien entre les entités et la base de données physique est possible grâce au Entity Data Model (EDM)



## La classe `DbContext`

Votre classe qui hérite de la classe de base DbContext est c’elle qui s’occupe des interactions avec la base de données.

Converti en objet .NET le data venant de base de données

Fait le suivi des modifications effectuer aux objets .NET

Persiste l’information des objets .NET sur la base de données à l’aide de la méthode:
- `SaveChanges()` : Exécute toutes les modifications au niveau de la base de données

## Le model et les relations conceptuelles avec EF

Le rôle des « Navigation Properties » est de représenter les relations conceptuelles, les « JOIN » basées sur les clés étrangères.
- 1 : 1 (One-to-one)
- 1 : N (One-to-many)
- N : 1 (Many-to-one)
- N : N (Many-to-many)



## Les outils

### Package Management Console
1. `Install-Package Microsoft.EntityFrameworkCore.Tools`
1. `Add-Migration maPremiereMigration`
1. `Update-Database`


### CLI

1. `dotnet add package Microsoft.EntityFrameworkCore.Design`
1. `dotnet ef migrations add maPremiereMigration`
1. `dotnet ef database update`
