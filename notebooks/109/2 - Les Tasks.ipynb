## Les Tasks

Depuis .NET 4.5, nous pouvons utiliser les mots clés:

- `async`: spécifie qu'une méthode, une expression lambda ou une méthode anonyme est asynchrone.
- `await`: interrompre l'exécution de la méthode jusqu'à ce que la tâche attendue se termine et retourne le contrôle au code appelant.


Les méthodes (méthode / expression lamda / méthode asynchrone) peuvent être déclarées avec le mot-clé `async` afin de leur permettre d’effectuer une opération <span style="color: red;">non-bloquante</span>.

Une méthode déclarée avec le mot clé `async` exécutera de manière <span style="color: red;">bloquante</span> jusqu’à ce que le mot-clé `await` est rencontré.


Les méthodes (méthode / expression lamda / méthode asynchrone) déclarées avec le mot clé `async` peuvent avoir plusieurs contexte `await`.

L'opérateur `await` indique au compilateur que la méthode `async` ne peut pas continuer au-delà de ce point, tant que le processus asynchrone attendu n'est pas terminé. Entre-temps, le contrôle retourne à l'appelant de la méthode `async`.



Le type de retour est l'un des types suivants :

-`Task<TResult>` si votre méthode a une instruction de retour dans laquelle l'opérande a le type `TResult`.


-`Task` si votre méthode n'a aucune instruction de retour, ou si elle a une instruction de retour sans opérande.


Le mot-clé `await` semble caché le `Task<T>` et nous retourne la valeur voulue dans le type voulue.



Le nom d'une méthode `async`, par convention, se termine par un suffixe « Async ».
