# Indexeurs

L’opérateur d’index (aussi appelé un indexeur) permet d’accéder à une valeur particulière d’une classe basée sur l’index spécifié.

- Tout comme l’accès d’un élément à l’intérieur d’un tableau ou collection.

Cet opérateur doit être implémenté afin d’être utilisé; il n’y a pas d’implémentation par défaut.

Il peut y avoir plus d’un opérateur d’index tant que le type de l’index est différent.

Les types les plus communs à utiliser pour l’index sont:
- `int`
- `string`


Un indexeur est constitué de:

- Modificateurs d’accès optionnels
- Type de retour
- Mot clé « `this` »
- Entre identificateur tableau (`[]`)
  - Le type de l’index
  - Le nom de variable de l’index
- Un accesseur
- Un mutateur



Tout comme une propriété, un indexeur peut être mis en mode lecture-seulement ou écriture-seulement.

Un indexeur peut contenir plus d’un index entre les identificateurs tableau.

- Afin de donner plus de précision pour la sélection
- Afin d’accéder des données multidimensionnelles
- Etc.
