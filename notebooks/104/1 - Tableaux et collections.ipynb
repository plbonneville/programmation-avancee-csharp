#### Tableau rectangulaire

Ce type de tableau est simple  utiliser puisque chaque range (2D) est de la mme taille.

Pour dclarer un tableau rectangulaire, il suffit de spcifier:
- Le type
- Lindicateur tableau contenant une virgule pour chaque dimension additionnelle

Pour initialiser un tableau rectangulaire vide, il suffit dutiliser:
- Le mot cl `new`
- Le type
- Le nombre dlments de chaque dimension entre lindicateur tableau, spar par une virgule



Pour initialiser un tableau rectangulaire avec des valeurs prdfinies, il suffit dutiliser:
- Le mot cl `new`
- Le type avec lindicateur tableau contenant une virgule pour chaque dimension additionnelle
- Les valeurs entre accolades et spares par des virgules
  - Chaque dimension est reprsente par un groupe additionnel daccolades et galement spare par des virgules



````
// ===============================================================
// Tableau rectangulaire initialise (3x4)
// ===============================================================

int[,] tableauRectangulaire2D = new int[,]
{
    {   8,  68,  10, -4 },
    {  66,   7, -74, -70},
    {-987, 100,  44,  93}
};
````

Pour accder une valeur dun tableau rectangulaire, il suffit de spcifier lindex de chaque dimension (spar par une virgule)  lintrieur dun indicateur tableau.



````
int[,] tableauRectangulaire2D = new int[3, 4];

for (int x = 0; x < 3; x++)
{
    for (int y = 0; y < 4; y++)
    {
        tableauRectangulaire2D[x, y] = x + 2 * y;
    }
}

for (int x = 0; x < 3; x++)
{
    for (int y = 0; y < 4; y++)
    {
        int valeur = tableauRectangulaire2D[x, y];
        Console.Write("{0}\t", valeur);
    }
    Console.WriteLine();
}
````

````
// Toutes dimensions sont perdues.
// Le 'for' est beaucoup mieux pour les tableaux a multiple dimensions.

foreach (var valeur in tableauRectangulaire2D)
{
    Console.WriteLine(valeur);
}
````

````
// Comment connaitre la longueur?
Console.WriteLine(tableauRectangulaire2D.Length); // Longueur = x * y

Console.WriteLine(tableauRectangulaire2D.GetLength(0)); // Longueur de x
Console.WriteLine(tableauRectangulaire2D.GetLength(1)); // Longueur de y

Console.WriteLine(tableauRectangulaire2D.Rank); // Nombre de dimensions
````

````
// Normalement les tableaux sont "zero based"...
// Mais il est possible de dclarer des tableaux commencent  1, 2, 3, etc (genre VB6)
// Pas recomend, sauf quand on doit 'parler' avec d'autres languages
Console.WriteLine("de {0} a {1}",
    tableauRectangulaire2D.GetLowerBound(0),
    tableauRectangulaire2D.GetUpperBound(0));
````

````
// ===============================================================
// Tableau rectangulaire vide (4x4x2)
// ===============================================================
int[, ,] tableauRectangulaire3D = new int[4, 4, 2];

// Initialisation du tableau rectangulaire
int c = 0;
for (int k = 0; k < tableauRectangulaire3D.GetLength(2); k++)
    for (int i = 0; i < tableauRectangulaire3D.GetLength(0); i++)
        for (int j = 0; j < tableauRectangulaire3D.GetLength(1); j++)
            tableauRectangulaire3D[i, j, k] = c++;

c = 0;
for (int feuille = 0; feuille < tableauRectangulaire3D.GetLength(2); feuille++)
    for (int x = 0; x < tableauRectangulaire3D.GetLength(0); x++)
        for (int y = 0; y < tableauRectangulaire3D.GetLength(1); y++)
            tableauRectangulaire3D[x, y, feuille] = c++;

/*
 * 0  4  8 13
 * 1  5  9 14
 * 2  6 10 15
 * 3  7 11 16
 * 
 * 
 * 17 21 25 29
 * 18 21 26 30
 * 19 23 27 31
 * 20 24 28 32
 */
````

````
// Affiche les valeurs du tableau 2D
Console.WriteLine("Tableau 2D");
for (int x = 0; x < tableauRectangulaire2D.GetLength(0); x++)
{
    for (int y = 0; y < tableauRectangulaire2D.GetLength(1); y++)
        Console.Write("{0}\t", tableauRectangulaire2D[x, y]);

    Console.WriteLine();
}
Console.WriteLine();

// Affiche les valeurs du tableau 3D
Console.WriteLine("Tableau 3D");
for (int z = 0; z < tableauRectangulaire3D.GetLength(2); z++)
{
    Console.WriteLine("[Z = {0}]", z);

    for (int x = 0; x < tableauRectangulaire3D.GetLength(0); x++)
    {
        for (int y = 0; y < tableauRectangulaire3D.GetLength(1); y++)
            Console.Write("{0}\t", tableauRectangulaire3D[x, y, z]);

        Console.WriteLine();
    }

    Console.WriteLine();
}
````

#### Tableau en escalier

Ce type de tableau est plus complexe  utiliser puisque chaque colonne contient une autre instance de tableau.

  > Puisque chaque colonne  son propre tableau, la taille des ranges (2D) ne sera pas ncessairement identique.

Pour dclarer un tableau en escalier, il suffit de spcifier:
- Le type
- Un indicateur tableau pour chaque dimension



Pour initialiser (la premire dimension) un tableau en escalier vide, il suffit dutiliser:

- Le mot cl `new`
- Le type
- Le nombre dlments dans la premire dimension entre un indicateur tableau
- Un indicateur tableau pour chaque dimension additionnelle



Chaque dimension subsquente doit tre initialise pour chaque range de la mme manire que la premire, mais en enlevant une dimension  chaque fois.



````
int[][] tableauEnEscalier2D = new int[3][];

for (int x = 0; x < 3; x++)
{
    for (int y = 0; y < 4; y++)
    {
        tableauEnEscalier2D[x][y] = x + 2 * y;
    }
}
````

````
int[][] tableauEnEscalier2D = new int[3][];

for (int x = 0; x < 3; x++)
{
    tableauEnEscalier2D[x] = new int[4]; // Obligatoire d'initialiser les tableaux dans le tableau.
    for (int y = 0; y < 4; y++)
    {
        tableauEnEscalier2D[x][y] = x + 2 * y;
    }
}

for (int x = 0; x < 3; x++)
{
    for (int y = 0; y < 4; y++)
    {
        int valeur = tableauRectangulaire2D[x, y];
        Console.Write("{0}\t", valeur);
    }
    Console.WriteLine();
}
````

## Collections

.NET comprend deux espaces de nom (namespace) spcialises pour des collections.

- Les collections non-gnriques : `System.Collections`
- Les collections gnriques : `System.Collections.Generic`



Afin daccder  ces collections, lespace de nom peut tre manuellement utilise ou peut tre rajoute au dbut du fichier avec les autres dclarations despaces de nom.


### Collections non-gnriques

Les collections non-gnriques font encore parties du .NET Framework 4.0  des fins de support pour des anciennes applications.

  > Les collections non-gnriques ne devraient pas tre utilises en .NET +2.0.
  
La plupart de ces collections ne sont pas fortement types; ils utilisent le type de base `object`.


Les collections non-gnriques implmentent toute linterface `ICollection`.

Linterface `ICollection` ajoute les membres suivants:
- `Count`: proprit reprsentant le nombre dlments dans la collection
- `CopyTo`: mthode copiant les lments de la collection  un tableau quelconque

Chaque collection possde des mthodes et champs membres spcifiques permettant lajout, le retrait et dautres fonctions particulires.


### Collections gnriques

Les collections gnriques sont prfres  leurs contreparties non gnriques puisquelles sont fortement types.

La dclaration et linitialisation dune collection gnrique exigent que le type d'objet  stocker est dfini.



Une collection gnrique peut tre initialise avec des valeurs prdfinies tout comme un tableau.



La plupart des collections supporte galement linitialisation de valeurs  partir dun tableau du mme type  laide dun constructeur spcialis.



Les collections gnriques implmentent toute linterface `ICollection<T>` (o `T` reprsente le type stock).
    
Linterface ICollection<T> ajoute les membres suivants:
- `Count`: proprit reprsentant le nombre dlments dans la collection
- `CopyTo`: mthode copiant les lments de la collection  un tableau quelconque
- `Add`: mthode ajoutant un lment  la collection
- `Clear`: mthode effaant tous les lments de la collection
- `Contains`: mthode indiquant si llment particulier existe dans la collection
- `Remove`: mthode supprimant un lment de la collection

    Chaque collection possde des mthodes et champs membres spcifiques permettant lajout, le retrait et dautres fonctions particulires.


````
ICollection<string> jours = new[] { "dimanche", "lundi", "mardi", "mercredi", "jeudi", "vendredi", "samedi" };
    
if (!jours.IsReadOnly)
    jours.Add("8e jour!");
else
    Console.WriteLine("La collection est en lecture seul");

foreach (var jour in jours)
{
    Console.WriteLine(jour);
}
````

#### `Queue<T>`

Les mthodes les plus utilises sont:

Une boucle `while` ou `do-while` peut tre utilise pour numrer (et vider) les lments de la collection.



````
// ===============================================================
// FIFO, pensez a une line a la banque
// ===============================================================

Queue<string> file = new Queue<string>();
file.Enqueue("abc");
file.Enqueue("def");
file.Enqueue("xyz");
file.Enqueue("123");

while (file.Count > 0)
{
    Console.WriteLine("Peek: " + file.Peek());
    Console.WriteLine("Peek: " + file.Peek());
    Console.WriteLine("Dequeue: " + file.Dequeue());
    //Console.WriteLine("Peek: " + file.Peek()); // System.InvalidOperationException: Queue empty.

    Console.WriteLine();
}
````

````
public record Personne(string Prenom, string Nom, int Age);

public void GetCoffee(Personne p)
{
    Console.WriteLine("{0} a eu son cafe!", p.Prenom);
}

// Faire un file "queue" avec trois personnes.
Queue<Personne> personneQ = new Queue<Personne>();
personneQ.Enqueue(new Personne ("Homer", "Simpson", 47));
personneQ.Enqueue(new Personne ("Marge", "Simpson", 45));
personneQ.Enqueue(new Personne ("Lisa",  "Simpson",  9));

// Qui est la premiere personne dans la file "queue"?
Console.WriteLine("{0} est la premiere personne en ligne!", personneQ.Peek().Prenom);

// Retirer chaque personne de la file "queue". 
GetCoffee(personneQ.Dequeue());
GetCoffee(personneQ.Dequeue());
GetCoffee(personneQ.Dequeue());

// Essayer de retirer une autre personne de la file "queue". ?
try
{
    GetCoffee(personneQ.Dequeue());
}
catch (InvalidOperationException e)
{
    Console.WriteLine("Erreur! {0}", e.Message);
}
````

#### `ConcurrentDictionary<TKey, TValue>`

Les mthodes les plus utilises sont:

- `TryAdd(key, value)`
- `TryUpdate(key, newValue, comparisonValue)`
- `AddOrUpdate(key, addValue, updateValueFactory)`
- `GetOrAdd(key, value)`

Attention, les accs aux lments sont beaucoup plus lents.


````
var d = new Dictionary<string, int>();
var sw = System.Diagnostics.Stopwatch.StartNew();
const int denominateur = 3;
const int total = 1_000_000;

foreach (var nombre in Enumerable.Range(0, total))
{
    if (nombre % denominateur == 0)
    {
        if (!d.ContainsKey(denominateur.ToString()))
        {
            // Ajoutons le premier lment
            d.Add(denominateur.ToString(), 1);
        }
        else
        {
            // Si l'lment existe ajoutons 1
            d[denominateur.ToString()]++;
        }
    }
}

Console.WriteLine("Il y a {0} nombre divisible par le denominateur {1} dans de 0 a {2}",
    d[denominateur.ToString()],
    denominateur,
    total);

sw.Stop();
Console.WriteLine("Temps total: {0}", sw.ElapsedMilliseconds);
````

````
using System.Collections.Concurrent;

var d = new ConcurrentDictionary<string, int>();
var sw = System.Diagnostics.Stopwatch.StartNew();
const int denominateur = 3;
const int total = 1_000_000;

System.Threading.Tasks.Parallel.ForEach(Enumerable.Range(0, total), (nombre) =>
{
    if (nombre % denominateur == 0)
    {
        d.AddOrUpdate(
            key: denominateur.ToString(),
            addValue: 1,
            updateValueFactory: (key, count) => { return count + 1; });
    }
}
);

Console.WriteLine("Il y a {0} nombre divisible par le denominateur {1} dans de 0 a {2}",
    d[denominateur.ToString()],
    denominateur,
    total);

sw.Stop();
Console.WriteLine("Temps total: {0}", sw.ElapsedMilliseconds);
````

## Covariance et contravariance

### Les enumerable et la covariance

Un tableau de type `string[]` peut tre converti en `object[]`.

Une liste `List<string>` ne peut pas tre converti en `List<object>` mais peut etre converti en `IEnumerable<object>` car `Ienumerable<T>` est en mode lecture seul.


### Covariance

Prenons: `IEnumerable<out T>`

Nous pouvons convertir un `IEnumerable<string>` en `IEnumerable<object>` car nous faisons que sortir des valeurs et nous obtenons un type spcifique, dans ce cas un `string`, et nous pouvons convertir ce type en objet plus gnral, dans ce cas un `object`.

### Contravariance

Prenons: `IComparer<in T>`

Nous pouvons convertir un `IComparer<object>` en un `IComparer<string>`, car les valeurs ne font quentrer et nous nous attendons  un type gnral, dans notre cas un `object`, et nous lui passons un type plus spcifique, dans notre cas un `string`.
